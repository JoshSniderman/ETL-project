```python
import requests
import pandas as pd
import numpy as np
```

```python
data_file = 'CSVs/Final_DataFrame.csv'
df = pd.read_csv(data_file)
```

```python
df
```

```python
url = "https://api.geocod.io/v1.6/geocode"

address_lat = []
address_lng = []
address_name = []

for i in range(0,1000):
    try:
        address_dummy = df["Address"][i].replace(" ", "+")
        city_dummy = df["City"][i]
    
        response = requests.get(f"{url}?q={address_dummy}+%2c+{city_dummy}&api_key=00822cf80e63d6daa666f6c9c90f0a029d3f8d3").json()
        address_lat.append(response["results"][0]["location"]["lat"])
        address_lng.append(response["results"][0]["location"]["lng"])
        address_name.append(df["Address"][i])
        print("coords added")
    except:
        print(f"City not found. Skipping...")
```

```python
coor_df = pd.DataFrame({
    "Address": address_name,
    "Lat": address_lat,
    "Lng": address_lng
})
```

```python
coor_df.to_csv("Coor_CSV.csv")
```
